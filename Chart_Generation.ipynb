```python
import os
import matplotlib.pyplot as plt

# Path to the dataset folder
input_folder = '/content/drive/MyDrive/MiniketRice/Dataset/Black_Background'

# Count the number of images in each class
class_counts = {}
for class_folder in os.listdir(input_folder):
    class_path = os.path.join(input_folder, class_folder)
    if os.path.isdir(class_path):
        class_counts[class_folder] = len([img for img in os.listdir(class_path) if img.endswith(('.png', '.jpg', '.jpeg'))])

# Data for the chart
classes = list(class_counts.keys())
counts = list(class_counts.values())

# Replicate the style of the chart
plt.figure(figsize=(12, 6))
bars = plt.bar(classes, counts, color='steelblue', edgecolor='black')  # Changed to dark cyan
plt.title('Original Images Class Distribution', fontsize=14)
plt.xlabel('Classes', fontsize=12)
plt.ylabel('Number of Images', fontsize=12)

# Add gridlines
plt.grid(axis='y', color='gray', linestyle='--', linewidth=0.5, alpha=0.5)  # Faint gray gridlines on y-axis

# Adjust tick sizes
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Tight layout for a clean look
plt.tight_layout()
plt.show()
```
